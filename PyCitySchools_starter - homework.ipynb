```python
totalstudentsperschool=school_data_complete["school_name"].value_counts()
totalschoolbudget=school_data_complete.groupby(["school_name"]).sum()["budget"]
schoolbudgetperstudent=totalschoolbudget/totalstudentsperschool
avgmathscoreperschool=school_data_complete.groupby(["school_name"]).mean()["math_score"]
avgreadingscoreperschool=school_data_complete.groupby(["school_name"]).mean()["reading_score"]

totalpassingmath=school_data_complete[school_data_complete["math_score"]>70]

totalpassingreading=school_data_complete[school_data_complete["reading_score"]>70]

perpassreadingperschool=totalpassingreading.groupby(["school_name"]).count()["student_name"]/totalstudentsperschool*100

perpassmathperschool=totalpassingmath.groupby(["school_name"]).count()["student_name"]/totalstudentsperschool*100

overallpassingrate=(perpassreadingperschool+perpassmathperschool)/2 
```
